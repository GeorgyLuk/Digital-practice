Конечно, это не спасёт нас в том случае, когда нам нужны и те и те кавычки. Но ЯП (языки программирования) проектируют тоже неглупые люди, и для решения этой проблемы существует подход, который называется *экранированием*. \
Откуда вообще берётся проблема? Дело в том, что кавычки, с одной стороны должны указывать питону, что то, что заключено внутри них - это строка. С другой стороны, кавычки могут быть смысловой частью строки. То есть как бы происходит смешение естественного (например, английского) и формального (языка программирования) языков. \
Так вот экранирование - это когда у нас есть специальный символ (чаще всего это обратный слэш, backslash - `\`), который можно поставить перед другим символом, чтобы обозначить, что этот символ (который идёт после `\`) должен быть именно текстом. В том смысле, что он не означает ничего специального, например, окончания строки.

```python
print('It's my life')
```

```python
print('It\'s very "simple"')  # по подсветке синтаксиса уже видно, что что-то не так
```

В ячейке выше как раз и показан пример того, как питон воспринимает кавычки, которые мы хотели для использовать для передачи сарказма, как кавычки, которые указывают на то, что строка заканчивается после `It is very `. А дальше, с точки зрения питона, написан какой-то бред, поэтому вас и выпадает `SyntaxError` - ошибка синтаксиса, то есть нарушения правил языка.

```python
# используем \ чтобы исправить ситуацию:
print("It is very \"simple\"")  # обратите внимание, обе "неправильные" кавычки необходимо экранировать
```

### Задача 1.
Дана последовательность ДНК:
short_dna = 'ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT'
Вам нужно написать код, который рассчитывает GC-состав - сколько процентов из последовательности составляют G и C нуклеотиды.
Почему это может быть нужно (для интереса): https://ru.wikipedia.org/wiki/GC-%D1%81%D0%BE%D1%81%D1%82%D0%B0%D0%B2

Вам нужно найти количество G нуклеотидов в сиквенсе (последовательности), количество C нуклеотидов. Для этого отлично подходит метод `.count()` \
Далее найти их сумму и просо эту сумму поделить на общее число нуклеотидов.

Вам нужно в коде ниже заменить все `pass` на код таким образом, чтобы вычислить GC-состав.

```python
short_dna = 'ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCATGGG'
g_number = short_dna.count("G")
c_number = short_dna.count("C")
overall_len = len(short_dna)
gc_content = (g_number + c_number) / overall_len

print("GC content is: ", gc_content)

print("GC content is: ", gc_content * 100, "%")  # то же самое, только в процентах
```

```python
gen_dna = "ATCGATCGATCGATCGACTGACTAGTCATAGCTATGCATGTAGCTACTCGATCGATCGATCGATCGATCGATCGATCGATCGATCATGCTATCATCGATCGATATCGATGCATCGACTACTAT"
exon1 = gen_dna[:63]
exon2 = gen_dna[90:]
intron = gen_dna[63:90]

print("Coding part: ", exon1 + exon2)
print("Intron is: ", intron)
```

#### 3.2
Подсчитайте соотношение количетсва нуклеотидов в интроне к количеству нуклеотидов в кодирующей части и выведите это соотношение в процентах.

```python
countintron = len(intron)
codingcoun = len(exon1 + exon2)
unserresult = countintron / codingcoun * 100


print(unserresult, '%')

```
